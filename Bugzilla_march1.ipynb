```python
def removeSpecialChars(df):
    '''
    Removes special characters which are specifically found in tweets.
    '''
    #Converts HTML tags to the characters they represent
    soup = BeautifulSoup(df, "html.parser")
    data = soup.get_text()

    #Convert www.* or https?://* to empty strings
    data = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',data)

    #Convert @username to empty strings
    data = re.sub('@[^\s]+','',data)
    
    #remove org.apache. like texts
    data =  re.sub('(\w+\.){2,}','',data)

    #Remove additional white spaces
    data = re.sub('[\s]+', ' ', data)
    
    data = re.sub('\.(?!$)', '', data)

    #Replace #word with word
    data = re.sub(r'#([^\s]+)', r'\1', data)

    return data 
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

nltk.download('averaged_perceptron_tagger')

# Define a function to extract nouns from a text string
def extract_nouns(text):
    tokens = word_tokenize(text)
    tagged = nltk.pos_tag(tokens)
    nouns = [word for word, pos in tagged if pos.startswith('N')]
    return ' '.join(nouns)

# Create a new 'tags' column by applying the extract_nouns function to the 'summary' column
df['summary_new'] = df['summary'].apply(extract_nouns)

# Display the resulting DataFrame
df
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Define the cluster number you want to analyze
cluster_num = 1

# Filter the dataset to keep only the rows belonging to the specified cluster
cluster_df = df[df['summary_new'] == cluster_num]

# Define a lemmatizer function
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(w) for w in nltk.word_tokenize(text)])

# Apply the lemmatizer function to the 'summary' column
cluster_df['lemmatized_summary'] = cluster_df['summary'].apply(lemmatize_text)

# Transform the lemmatized 'summary' column into a vector representation using TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(cluster_df['lemmatized_summary'])

# Train a NearestNeighbors model on the tf-idf matrix
k = 5  # number of neighbors to consider
nearest_neighbors = NearestNeighbors(n_neighbors=k, metric='cosine').fit(tfidf_matrix)

# Find the k most similar rows for each row in the dataset
distances, indices = nearest_neighbors.kneighbors(tfidf_matrix)

# Convert the indices to row labels
row_labels = [cluster_df.index[i] for i in indices.flatten()]

# Create a DataFrame with the most similar rows
most_similar_df = pd.DataFrame({
    'similarity_score': distances.flatten(),
    'most_similar_row': row_labels
})

# Group the DataFrame by 'most_similar_row' and count the number of occurrences
count_df = most_similar_df.groupby('most_similar_row').count()

# Sort the DataFrame by the count of occurrences, in descending order
count_df = count_df.sort_values(by='similarity_score', ascending=False)

# Join the count DataFrame with the original DataFrame to get the full rows
output_df = pd.merge(count_df, df, left_index=True, right_index=True, how='inner')

# Sort the DataFrame by the count of occurrences again, in descending order
output_df = output_df.sort_values(by='similarity_score', ascending=False)

# Print the output DataFrame
print(output_df)
```
